```python
# Function to Display Image
def displayImage(image):
    if len(image.shape)==3:
        color_coverted = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        plt.imshow(color_coverted)
        plt.show()
        
    else:
        plt.imshow(image, cmap="gray")
        plt.show()

  # initialize the Haar Cascade face detection model
face_cascade = cv2.CascadeClassifier(cv2.samples.findFile(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'))
```

```python
# Open image using openCV2
img1 = cv2.imread(path+"smile.jpg")
img2 = cv2.imread(path+"angry.jpg")
img3 = cv2.imread(path+"normal.jpg")
color_coverted = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
color_coverted2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
color_coverted3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

#displayImage(img1)
#displayImage(img2)
#displayImage(img3)
plt.figure(figsize=(10,3))
plt.subplot(131); plt.title("Smile"); plt.imshow(color_coverted, cmap="gray");
plt.subplot(132); plt.title("Angry"); plt.imshow(color_coverted2, cmap="gray");
plt.subplot(133); plt.title("Normal"); plt.imshow(color_coverted3, cmap="gray");
```

```python
img1 = cv2.imread(path+"smile.jpg")
img2 = cv2.imread(path+"angry.jpg")
img3 = cv2.imread(path+"normal.jpg")


faces = face_cascade.detectMultiScale(img1,1.3,3)
  # draw face bounding box on image1
for (x,y,w,h) in faces:
    cv2.rectangle(img1,(x,y),(x+w,y+h),(0,255,0),2)
    roi = img1[y:y+h,x:x+w]
    roi = cv2.Canny(roi, 23, 30)
    img1[y:y+roi.shape[0], x:x+roi.shape[1]] = np.expand_dims(roi, axis = -1)
    

faces = face_cascade.detectMultiScale(img2,1.3,3)
  # draw face bounding box on image2
for (x,y,w,h) in faces:
    cv2.rectangle(img2,(x,y),(x+w,y+h),(0,255,0),2)
    roi = img2[y:y+h,x:x+w]
    roi = cv2.Canny(roi, 23, 30)
    img2[y:y+roi.shape[0], x:x+roi.shape[1]] = np.expand_dims(roi, axis = -1)
    

faces = face_cascade.detectMultiScale(img3,1.3,3)
  # draw face bounding box on image3
for (x,y,w,h) in faces:
    cv2.rectangle(img3,(x,y),(x+w,y+h),(0,255,0),2)
    roi = img3[y:y+h,x:x+w]
    roi = cv2.Canny(roi, 23, 30)
    img3[y:y+roi.shape[0], x:x+roi.shape[1]] = np.expand_dims(roi, axis = -1)
    


displayImage(img1)
displayImage(img2)
displayImage(img3)

# displayImage(imgSobel)
# displayImage(imgCanny2)
# displayImage(imgSobel2)
# displayImage(imgCanny3)
# displayImage(imgSobel3)
```

```python
img1 = cv2.imread(path+"smile.jpg")
img2 = cv2.imread(path+"angry.jpg")
img3 = cv2.imread(path+"normal.jpg")


faces = face_cascade.detectMultiScale(img1,1.3,3)
  # draw face bounding box on image1
for (x,y,w,h) in faces:
    cv2.rectangle(img1,(x,y),(x+w,y+h),(0,255,0),2)
    roi = img1[y:y+h,x:x+w]
    sobelx = cv2.Sobel(roi, cv2.CV_8U,1,0,ksize=3)
    sobely = cv2.Sobel(roi, cv2.CV_8U,0,1,ksize=3)
    roi = sobelx + sobely
    img1[y:y+roi.shape[0], x:x+roi.shape[1]] = np.expand_dims(roi, axis = 0)

faces = face_cascade.detectMultiScale(img2,1.3,3)
  # draw face bounding box on image1
for (x,y,w,h) in faces:
    cv2.rectangle(img2,(x,y),(x+w,y+h),(0,255,0),2)
    roi = img2[y:y+h,x:x+w]
    sobelx = cv2.Sobel(roi, cv2.CV_8U,1,0,ksize=3)
    sobely = cv2.Sobel(roi, cv2.CV_8U,0,1,ksize=3)
    roi = sobelx + sobely
    img2[y:y+roi.shape[0], x:x+roi.shape[1]] = np.expand_dims(roi, axis = 0)

faces = face_cascade.detectMultiScale(img3,1.3,3)
  # draw face bounding box on image1
for (x,y,w,h) in faces:
    cv2.rectangle(img3,(x,y),(x+w,y+h),(0,255,0),2)
    roi = img3[y:y+h,x:x+w]
    sobelx = cv2.Sobel(roi, cv2.CV_8U,1,0,ksize=3)
    sobely = cv2.Sobel(roi, cv2.CV_8U,0,1,ksize=3)
    roi = sobelx + sobely
    img3[y:y+roi.shape[0], x:x+roi.shape[1]] = np.expand_dims(roi, axis = 0)

displayImage(img1)
displayImage(img2)
displayImage(img3)
```

```python
img1 = cv2.imread(path+"smile.jpg")
img2 = cv2.imread(path+"angry.jpg")
img3 = cv2.imread(path+"normal.jpg")

color_coverted = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
color_coverted2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
color_coverted3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

faces = face_cascade.detectMultiScale(color_coverted)
  # draw face bounding box on image1
for (x,y,w,h) in faces:
    color_coverted[y:y+h,x:x+w] = cv2.medianBlur(color_coverted[y:y+h,x:x+w],35)

faces = face_cascade.detectMultiScale(color_coverted2)
  # draw face bounding box on image2
for (x,y,w,h) in faces:
    color_coverted2[y:y+h,x:x+w] = cv2.medianBlur(color_coverted2[y:y+h,x:x+w],35)

faces = face_cascade.detectMultiScale(color_coverted3)
  # draw face bounding box on image3
for (x,y,w,h) in faces:
    color_coverted3[y:y+h,x:x+w] = cv2.medianBlur(color_coverted3[y:y+h,x:x+w],35)


plt.subplot(131); plt.title("Smile"); plt.imshow(color_coverted)
plt.subplot(132); plt.title("Angry"); plt.imshow(color_coverted2)
plt.subplot(133); plt.title("Normal"); plt.imshow(color_coverted3)
```

```python
path = "/content/gdrive/MyDrive/Assignment/"
image = cv2.imread(path+'coins.jpg')

coin_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

coin_blur = cv2.GaussianBlur(coin_gray, (11,11), 0) 

coin_canny = cv2.Canny(coin_blur, 5,150,3)

coin_dilated = cv2.dilate(coin_canny,(1,1), iterations = 20)

(countour, hierarchy) = cv2.findContours(coin_dilated.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
coin_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
cv2.drawContours(coin_rgb,countour, -1, (0,255,0),2)

print(len(countour))

plt.figure(figsize=(15,3))
plt.subplot(161); plt.title("Original"); plt.imshow(image)
plt.subplot(162); plt.title("Gray"); plt.imshow(coin_gray, cmap='gray')
plt.subplot(163); plt.title("Blur"); plt.imshow(coin_blur, cmap='gray')
plt.subplot(164); plt.title("Canny"); plt.imshow(coin_canny, cmap='gray')
plt.subplot(165); plt.title("Dilated"); plt.imshow(coin_dilated, cmap='gray')
plt.subplot(166); plt.title("Processed"); plt.imshow(coin_rgb)
plt.show()
```
